```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import celluloid as cld
import seaborn as sns
```

```python
sequential_data = pd.read_csv('output_1_thread.csv', sep=',')
sequential_data
```

```python
parallel_data_8_thr = pd.read_csv('output_8_thread.csv', sep=',')
parallel_data_8_thr
```

```python
parallel_data_16_thr = pd.read_csv('output_16_thread.csv', sep=',')
parallel_data_16_thr
```

```python
parallel_data_32_thr = pd.read_csv('output_32_thread.csv', sep=',')
parallel_data_32_thr
```

# Calculation time

```python
plt.plot(sequential_data.t, sequential_data.Time)
plt.xlabel('step')
plt.ylabel('calculation time')
plt.title('Execution time graph when sequential')

plt.show()
```

```python
plt.plot(parallel_data_8_thr.t, parallel_data_8_thr.Time)
plt.xlabel('step')
plt.ylabel('calculation time')
plt.title('Execution time graph when parallel (8 threads)')

plt.show()
```

```python
plt.plot(parallel_data_16_thr.t, parallel_data_16_thr.Time)
plt.xlabel('step')
plt.ylabel('calculation time')
plt.title('Execution time graph when parallel (16 threads)')

plt.show()
```

```python
plt.plot(parallel_data_32_thr.t, parallel_data_32_thr.Time)
plt.xlabel('step')
plt.ylabel('calculation time')
plt.title('Execution time graph when parallel (32 threads)')

plt.show()
```

# Speedups

```python
fig, ax = plt.subplots(figsize=(20, 5))
sns.lineplot(x=parallel_data_8_thr.index, y=(sequential_data.Time / parallel_data_8_thr.Time), ax=ax)
plt.title("Speedup between 1 and 8 threads")
plt.xlabel("Step")
plt.ylabel("Calculation time")
```

```python
fig, ax = plt.subplots(figsize=(20, 5))
sns.lineplot(x=parallel_data_16_thr.index, y=(sequential_data.Time / parallel_data_16_thr.Time), ax=ax)
plt.title("Speedup between 1 and 16 threads")
plt.xlabel("Step")
plt.ylabel("Calculation time")
```

```python
fig, ax = plt.subplots(figsize=(20, 5))
sns.lineplot(x=parallel_data_32_thr.index, y=(sequential_data.Time / parallel_data_32_thr.Time), ax=ax)
plt.title("Speedup between 1 and 32 threads")
plt.xlabel("Step")
plt.ylabel("Calculation time")
```

Efficiency equals to $E = \frac{S}{p}$ where S - time in parallel computing, p - number of threads

```python
def make_movie(df: pd.DataFrame, fileName: str):
    camera = cld.Camera(plt.figure())
    for step in range(2000):    #len(df)):
        plt.title(fileName)
        ax = plt.gca().set_aspect('equal')
        x_val = []
        y_val = []
        for i in range(1,51):
            x_val.append(df[f'x_{i}'][step])
            y_val.append(df[f'y_{i}'][step])
        plt.scatter(x=x_val, y=y_val, c='b', s=30)
        camera.snap()
    anim = camera.animate(blit=True)
    anim.save(fileName + ".gif")
```

```python
make_movie(parallel_data_32_thr, 'animation.gif')
# make_movie(sequential_data, 'animation2.gif')
```
